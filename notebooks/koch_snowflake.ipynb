```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess

%matplotlib inline
```

```python
current = 'f--f--f'
axiom = 'f'
rule = "f+f--f+f"
angle = 0
r = 1
plus_angle = np.pi/3
minus_angle = -np.pi/3
points = [0.+0.j]

threshold = 1_000_000

while len(current) <= threshold:
    current = current.replace(axiom, rule)
    print(len(current))
```

```python
for i in current:
    if i == "+":
        angle += plus_angle
    elif i == "-":
        angle += minus_angle
    elif i == "f":
        points.append(points[-1]+r*np.exp(1j*angle))
```

```python
name = "koch_snowflake.png"

plt.figure(figsize=(5,5), dpi=1200)
plt.axis("off")
plt.plot([p.real for p in points], [p.imag for p in points], linewidth=0.2)
plt.savefig("../images/"+name, bbox_inches='tight')
```
