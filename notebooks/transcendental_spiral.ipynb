```python
from mpmath import mp
import numpy as np
import matplotlib.pyplot as plt
import subprocess

%matplotlib inline
```

```python
prec = 1000
mp.dps = prec  # set number of digits
str_pi = str(mp.pi)
ints = '1234567890'
angle = 0
angle_inc = 1
scale = angle_inc * np.pi/5
points = [0.+0.j]
r = 1

num_points = 1_000_000

for i in str_pi:
        if i in ints:
                points.append(points[-1]+r*np.exp(1j*angle))
                angle += int(i)*scale
```

```python
name = "transcendental_spiral.png"

plt.figure(dpi=1200)
plt.axis("off")
plt.plot([p.real for p in points], [p.imag for p in points], linewidth=0.2)
plt.savefig("../images/"+name, bbox_inches='tight')
```
