```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess

%matplotlib inline
```

```python
a = 0.6+0.45j
points = [0.+0.j, 1.+0.j]
points = np.array(points)
itrs = 10
for i in range(itrs):
    f1 = a*np.conjugate(points)
    f2 = a + (1-a)*np.conjugate(points)
    points = np.hstack((f1[::-1],f2))
```

```python
name = "koch_peano_curve.png"

plt.figure(dpi=1200)
plt.axis("off")
plt.plot([p.real for p in points], [p.imag for p in points], linewidth=0.2)
plt.savefig("../images/"+name, bbox_inches='tight')
```
