```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess
import re

%matplotlib inline
```

```python
current = 'x'
rules = {
    'x':"xfyfx+f+yfxfy-f-xfyfx",
    'y':"yfxfy-f-xfyfx+f+yfxfy"
}
rep = dict((re.escape(k), v) for k, v in rules.items()) 
pattern = re.compile("|".join(rep.keys()))
angle = 0
r = 1
plus_angle = np.pi/2
minus_angle = -np.pi/2
points = [0.+0.j]

threshold = 100_000

while len(current) <= threshold:
    current = pattern.sub(lambda m: rep[re.escape(m.group(0))], current)
    print(len(current))
```

```python
for i in current:
    if i == "+":
        angle += plus_angle
    elif i == "-":
        angle += minus_angle
    elif i == "f":
        points.append(points[-1]+r*np.exp(1j*angle))
```

```python
name = "peano_curve.png"

plt.figure(figsize=(5,5), dpi=1200)
plt.axis("off")
plt.plot([p.real for p in points], [p.imag for p in points], linewidth=0.2)
plt.savefig("../images/"+name, bbox_inches='tight')
```
