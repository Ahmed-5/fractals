```python
import numpy as np
import matplotlib.pyplot as plt
import copy

%matplotlib inline
```

```python
init = np.array([[0,0], [0,1], [1,1], [1,0]])
```

```python
def rot90(x):
    d = x.max()
    n = copy.deepcopy(x)
    n = n[:,::-1]
    n[:,1] = d - n[:,1]
    n = n[::-1]
    return n

def rot270(x):
    d = x.max()
    n = copy.deepcopy(x)
    n = n[:,::-1]
    n[:,0] = d - n[:,0]
    n = n[::-1]
    return n
```

```python
order = 8
points = np.array([[0,0], [0,1], [1,1], [1,0]])
for i in range(1, order):
    p1 = copy.deepcopy(points)
    p2 = copy.deepcopy(points)
    p4 = copy.deepcopy(points)
    p3 = copy.deepcopy(points)
    # print(points)

    p1 = rot90(p1)
    p4 = rot270(p4)

    p1 = init[0]*(2**i) + p1
    p2 = init[1]*(2**i) + p2
    p3 = init[2]*(2**i) + p3
    p4 = init[3]*(2**i) + p4

    points = np.vstack((copy.deepcopy(p1), copy.deepcopy(p2), copy.deepcopy(p3), copy.deepcopy(p4)))
    # diff = 0
```

```python
plt.figure(figsize=(5,5), dpi=1200)
plt.axis("off")
plt.plot([p[0] for p in points], [p[1] for p in points], linewidth=0.2)
```
