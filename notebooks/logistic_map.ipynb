```python
from math import inf
import numpy as np
import matplotlib.pyplot as plt
import subprocess

%matplotlib inline
```

```python
xs = 0
xe = 4
ys = 0
ye = 1
x_p = 4001
y_p = 4001
x,y = np.meshgrid(np.linspace(xs, xe, x_p), np.linspace(ys, ye, y_p, endpoint=False)[1:])
x = x.ravel()
y = y.ravel()

inf_comp = np.array(complex(inf, inf))
nanj = np.complex128(np.nan+(np.nan)*1j)
```

```python
num_iter = 1000

for i in range(num_iter):
    y = x*y*(1-y)
```

```python
name = "logistic_map.png"

plt.figure(figsize=(4,1), dpi=1200)
plt.axis("off")
plt.scatter(x, y, s=1/720, linewidths=0.0001)
plt.savefig("../images/"+name, bbox_inches='tight')
```
