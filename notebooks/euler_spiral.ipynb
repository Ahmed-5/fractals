```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess

%matplotlib inline
```

```python
angle = 0
angle_inc = (1 + np.sqrt(5))/2
scale = angle_inc * np.pi/180
points = [0.+0.j]
r = 1

num_points = 1_000_000

for i in range(num_points):
        points.append(points[-1]+r*np.exp(1j*angle))
        angle += i*scale
```

```python
name = "euler_spiral.png"

plt.figure(dpi=1200)
plt.axis("off")
plt.plot([p.real for p in points], [p.imag for p in points], linewidth=0.2)
plt.savefig("../images/"+name, bbox_inches='tight')
```
