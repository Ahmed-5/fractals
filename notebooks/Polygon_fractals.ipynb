```python
import matplotlib.pyplot as plt
import numpy as np
from random import random, choice
from math import sqrt
```

```python
def mid_point(x1,y1,x2,y2):
    return (x1+x2)/2, (y1+y2)/2
```

```python
def gen_points(points, shape, itr):
    '''
    generate real sub-fractal of the polygon fractal
    inputs:
    points: (array of real number coordinates) the current sub-fractal
    shape: (array of real number coordinates) the starting shape used for the fractal
    itr: (int) index of the current itration
    '''
    l = len(points)
    for i in range(len(shape)**itr-1,l):
        for j in range(len(shape)):
            x,y = mid_point(points[i][0],points[i][1],shape[j][0],shape[j][1])
            points.append([x,y])

def chaos_game(n, shape):
    '''
    generate approximation of the polygon fractal
    inputs:
    n: (int) the number of points will be used to approximate the fractal
    shape: (array of real number coordinates) the starting shape used for the fractal
    '''
    point = [random(), random()]
    for i in range(n):
        point = [(p+s)/2 for p,s in zip(point, choice(shape))]
        yield point
```

```python
shape = [[0, 0], [0.5, sqrt(0.75)], [1, 0]]
points = [[0, 0], [0.5, sqrt(0.75)], [1, 0]]
for i in range(10):
    gen_points(points, shape, i)
print(len(points))
ps = []
for i in chaos_game(10000, shape):
    ps.append(i)
```

```python
plt.axis("off")
plt.scatter([i[0] for i in ps], [i[1] for i in ps], s=1)
```
