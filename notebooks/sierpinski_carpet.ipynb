```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess

%matplotlib inline
```

```python
offsets = [0.+0.j, 0.+1.j, 0.+2.j, 1.+0.j, 1.+2.j, 2.+0.j, 2.+1.j, 2.+2.j]
offsets = np.array(offsets)/3
```

```python
points = [0.+0.j, 1.+0.j, 1.+1.j, 0.+1.j]
points = np.array(points)
points = np.hstack((points, points/3 + (1.+1.j)/3))
itrs = 5
for i in range(itrs):
    points = points/3
    points = np.hstack((points + offset for offset in offsets))
```

```python
points.shape
```

```python
name = "sierpinski_carpet.png"

plt.figure(figsize=(5,5), dpi=1200)
plt.axis("off")
plt.scatter([p.real for p in points], [p.imag for p in points], s=1/72)
plt.savefig("../images/"+name, bbox_inches='tight')
```
